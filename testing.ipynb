# Set up

```python
import numpy as np
import pandas as pd
import streamlit as st
import plotly.express as px


df = pd.read_excel('Workout 2026.xlsx')[:6]
DOW = df.columns
DOW
```

# Getting the Daily Workouts

```python
Workouts_Per_Day = {}
for day in DOW:
    current_day_excecises = df[day]
    Workouts_Per_Day[day] = []


    #print(current_day_excecises.dropna())

    for item in current_day_excecises.dropna()[1:]:
        if item != "Relax - Stretch - Yoga":
            Workouts_Per_Day[day].append(pd.read_excel('Workout 2026.xlsx', sheet_name=item)['Excercise Name'][0])
    
Workouts_Per_Day
```

# Getting the Workout's Sheet from the .xlsx file

```python
day = 'Sunday'
# Excercise Name	Excercise Description	Date	Rest between Sets (Minutes)	Weight (lbs)	Sets (Count)	Reps (Count)	Volume (lbs)

for workout in Workouts_Per_Day[day]:
    day_df = pd.read_excel('Workout 2026.xlsx', sheet_name=workout)
    day_df
px.bar(data_frame=day_df,x='Date', y='Volume (lbs)', title=day_df['Excercise Name'][1])
```
